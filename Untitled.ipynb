```python
with open('static/hungarian_white_cards.txt', 'r', encoding='utf-8') as file:
    white_cards = [
        {
            'text' : line.replace('\n','').strip(),
            'icon' : 'hungarian',
            'deck' : 'hungarian'
        }
        
         for line in file.readlines() if not line.startswith('\n')]

def calc_blank(text):
    if text.count('BLANK') == 0:
        return 1
    else:
        return text.count('BLANK')
    
with open('static/hungarian_back_cards.txt', 'r', encoding = 'utf-8') as file:
    black_cards = [
        {
            'text' : line.replace('\n','').replace('BLANK', '_').strip(),
            'icon' : 'hungarian',
            'deck' : 'hungarian',
            'pick' : calc_blank(line.replace('\n',''))
        } for line in file.readlines() if not line.startswith('\n')]
```

```python
import json

with open('static/hungarian_cards.json', 'w') as outfile:
    data = {
        'white' : white_cards,
        'black' : black_cards
    }
    json.dump(data,  outfile ,indent=4)
```
